```python
#this is to check what my schema naming is
import os
from sqlalchemy import create_engine, inspect, text
from dotenv import load_dotenv

load_dotenv()

# Setup Connection
DB_URI = (
    f"postgresql://{os.getenv('PG_USER')}:{os.getenv('PG_PASSWORD')}@"
    f"{os.getenv('PG_HOST')}:{os.getenv('PG_PORT')}/{os.getenv('PG_DB')}"
)
engine = create_engine(DB_URI)

print(f"‚úÖ Connected to: {os.getenv('PG_DB')} @ {os.getenv('PG_HOST')}")

inspector = inspect(engine)
schemas = inspector.get_schema_names()

print("\nüîé FOUND SCHEMAS:")
print(schemas)

print("\nüîé SEARCHING FOR 'fct_image_detections'...")
found = False
for schema in schemas:
    try:
        tables = inspector.get_table_names(schema=schema)
        # Also check views just in case
        views = inspector.get_view_names(schema=schema)
        all_objects = tables + views

        if 'fct_image_detections' in all_objects:
            print(f"   üéâ FOUND IT! It is in schema: '{schema}'")
            print(f"      Full Name: {schema}.fct_image_detections")
            found = True
    except Exception as e:
        print(f"   Could not read schema {schema}: {e}")

if not found:
    print("‚ùå ERROR: Could not find 'fct_image_detections' in any schema.")
```
